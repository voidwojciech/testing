# Concrete database

## What is this database?
Data from papers is collected and presented in form of an interactive database. Instead of direct test results, it focuses on test type and conditions: storage conditions (water, air, duration); measured strength (post-fire residual, hot unstressed, hot stressed); heating rate (°C/min); target temperatures; duration of exposure and cooling methods. The database provides information whether specimens are oven-dried before putting them in a furnace, which is a common mean to mitigate spalling if polypropylene fibres are not present. Moreover, specimen shapes and sizes are given, since they range from 40 to 150 mm cubes or 36 to 150 mm cylinders. Materials are also listed, so database can be easily filtered for any research on, for example, High Alumina Cement or Fly Ash.

## Aim
The aim of this database is to help quickly find relevant references which might be helpful in planning tests or deciding whether some parameters might have a significant effect on the results.

## Motivation
When we were doing our research on properties of lightweight cement composites with cenospheres, we asked several questions:
- What are the commonly used specimen shapes and sizes, when we want to evaluate the residual compressive strength in per cent?
- What are the recommended heating and cooling rates (°C/min)? Is standard temperature-time curve being used in material testing?
- How common is it to dry specimens in an oven (usually at 105 °C) prior to heating?
The collected reference might be helpful when answering the question regarding the effect of one, or more parameters used in the test.

## Example 3a:
Show papers, where concrete was tested using $40 \ mm$ specimens AND temperatures $700$ OR $1000 \ °C$:<br>

**use $\&$ for AND operator; use $|$ for OR operator.**

```python
betons.loc[((betons['Specimen size (mm)'].str.contains('40')) & 
            (betons['Temperatures'].str.contains('700|1000')))]
```

## Example 3b:
