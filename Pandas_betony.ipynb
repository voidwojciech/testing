# Concrete database

## What is this database?
This interactive database presents data collected from several papers dealing with the effect that elevated temperature has on concrete.

Instead of direct test results, it focuses on conditions used during tests, such as:
- storage conditions (water, air, duration);
- specimen shape and size;
- measured type of strength (post-fire residual, hot unstressed, hot prestressed);
- heating rate (°C/min); 
- target temperatures;
- duration of heat exposure;
- cooling methods.

The database provides information, for example whether specimens are oven-dried before putting them in a furnace, which is a common mean to mitigate spalling if polypropylene fibres are not present. Moreover, specimen shapes and sizes are given, ranging from 25 to 150 mm cubes or 36 to 150 mm cylinders. Materials are also listed, so database can be easily filtered for research on, for example, High Alumina Cement or Fly Ash. If the paper contains additional test results, for example flexural strength, porosimetry, TGA or DSC, they are listed as well.

## Aim
The aim of this database is to help quickly find relevant references which might be helpful in planning tests or deciding whether some parameters might have a significant effect on the results.

## Motivation
During ongoing research on properties of lightweight cement composites with cenospheres, several questions araised:
- What are the commonly used specimen shapes and sizes, when we want to evaluate the residual compressive strength in per cent?
- What are the recommended heating and cooling rates (°C/min)? Is standard temperature-time curve being used in material testing?
- How common is it to dry specimens in an oven (usually at 105 °C) prior to heating?

The collected references might be helpful when answering questions regarding the effect of one or more parameters used in the test.

## Example 2:
Show papers, where concrete was exposed to 1200 °C:

```python
betons.loc[((betons['Temperatures'].str.contains('1200', na=False)))].style.format({'LINK': make_clickable})
# na=False: ignore nans: https://stackoverflow.com/questions/28311655/ignoring-nans-with-str-contains
```

## Example 3a:
Show papers, where concrete was tested using $40 \ mm$ specimens AND temperatures $700$ OR $1000 \ °C$:<br>

**use $\&$ for AND operator; use $|$ for OR operator.**

```python
betons.loc[((betons['Specimen size (mm)'].str.contains('40')) & 
            (betons['Temperatures'].str.contains('700|1000')))].style.format({'LINK': make_clickable})
```

## Example 3b:
Show papers, where concrete was tested using $40 \ mm$ specimens  and temperatures $700$ OR $1000 \ °C$, but show only selected columns: (*Specimen size (mm)*, *Temperatures* and *Cooling*)

```python
betons.loc[((betons['Specimen size (mm)'].str.contains('40')) & 
            (betons['Temperatures'].str.contains('700|1000'))), 
            ('Author', 'Specimen size (mm)', 'Test type', 'Cooling')].style.format({'LINK': make_clickable})
```

## Example 3c:
Show papers printed in 2018 considering *Heating rate* and *Test type*.

```python
betons.loc[((betons['Author'].str.contains('2018', na=False))), 
           ('Author', 'Heating rate', 'Test type')].style.format({'LINK': make_clickable})
```
